```python
# import the libraries needed
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score,precision_score,recall_score,f1_score
import xgboost as xgb
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

```

```python
data=pd.read_csv('customer_churn_prediction_dataset.csv')
```

```python
data.head()
```

```python
data.info()
```

```python
data=data.drop(['customerID'],axis=1)
```

```python
data.info()
```

# no need to import but do need to encode the data due to the fact that there are different types of data 



```python
# decide 
for col in data.columns:
    print(col, data[col].nunique())
int_encode_cols = [
    'gender',
    'Partner',
    'Dependents',
    'PhoneService',
    'PaperlessBilling',
    'Churn'
]

onehot_encode_cols = [
    'MultipleLines',
    'InternetService',
    'OnlineSecurity',
    'OnlineBackup',
    'DeviceProtection',
    'TechSupport',
    'StreamingTV',
    'StreamingMovies',
    'Contract',
    'PaymentMethod'
]

```

```python
from sklearn.preprocessing import LabelEncoder
import pandas as pd

#

encoder = LabelEncoder()
# copy the df just so I can double check quality
data_transformed = data.copy()

for col in int_encode_cols:
    data_transformed[col] = encoder.fit_transform(data[col])

data_transformed = pd.get_dummies(data_transformed, columns=onehot_encode_cols, drop_first=True)

```

```python
data_transformed.info()
```

# Visualizations we will use pair plot to determine which ones to train over


```python

# Separate features and target
features = data_transformed.drop(['Churn'], axis=1)
Y = data_transformed['Churn']

df = data_transformed.copy()
df['Churn'] = Y

# Make sure the pair plots are numeric they should
df_numeric = df.select_dtypes(include=['number'])

# Pair plot
sns.pairplot(
    df_numeric,
    hue='Churn',
    diag_kind='hist',
    palette='Set1'
)

plt.savefig("pairplot.png")

plt.show()
```

# Feature Selection: <---
- Notes there are way to manyfeature to samples 
- there are clas imbalances  so we can cc leant these out

# Now choose columns for target(Y) and features(X)

- Based on the visualizations above we can try and thin out not important data 

```python
Y = data_transformed['Churn']
X=data_transformed.drop(['Churn'],axis=1)
```

```python
X.shape
```

# Make the Decsion Tree Model 

```python
#Split into train/test
X_train, X_test, Y_train, Y_test = train_test_split(
    X, Y, test_size=0.2, random_state=42
)

#Initialize the Decision Tree classifier
dt = DecisionTreeClassifier(random_state=42)

#Train the model
dt.fit(X_train, Y_train)

#Make predictions
Y_pred = dt.predict(X_test)

# eval
print("Accuracy Score:\n", accuracy_score(Y_test,Y_pred))
print("Precision Score:\n", precision_score(Y_test,Y_pred))
print("Recall Score:\n", recall_score(Y_test,Y_pred))
print("F1 Score:\n", f1_score(Y_test,Y_pred))
print("Classification Report:\n", classification_report(Y_test, Y_pred))
#print("\nConfusion Matrix:\n", confusion_matrix(Y_test, Y_pred))
```

```python

from sklearn.tree import export_text

tree_rules = export_text(dt, feature_names=list(X.columns))
#print(tree_rules)


```

```python
importances = dt.feature_importances_
feature_names = X.columns

f_i_df = pd.DataFrame({
    "Feature": feature_names,
    "Gini Importance": importances
}).sort_values(by="Gini Importance", ascending=False)

print(f_i_df)
```

```python
rf = RandomForestClassifier(
    n_estimators=100,       
    max_depth=5,          
    random_state=42,
    class_weight='balanced',
    bootstrap=True
)

rf.fit(X_train, Y_train)

Y_pred_rf = rf.predict(X_test)

print("Accuracy Score:", accuracy_score(Y_test, Y_pred_rf))
print("Precision Score:", precision_score(Y_test, Y_pred_rf))
print("Recall Score:", recall_score(Y_test, Y_pred_rf))
print("F1 Score:", f1_score(Y_test, Y_pred_rf))
print("\nClassification Report:\n", classification_report(Y_test, Y_pred_rf))
```
